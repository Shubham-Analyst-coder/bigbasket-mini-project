```python
df.isnull().sum()
```

```python
product = df.groupby(['product',"sale_price"]).size().reset_index(name="count")
product
```

```python
fig=px.bar(product.head(30),x="product" , y="sale_price",title=" Top 30 product", color_discrete_sequence=['orange'])
fig.update_layout(
    title_x=0.5,
    xaxis_title="Product",
    yaxis_title="Sale Price",
    hovermode="x unified",
    xaxis={'categoryorder':'total descending'} # Order bars by sale price
)
fig.update_traces(hovertemplate='Product: %{x}<br>Sale Price: %{y}')
fig.show()
```

```python
df["market_price"].mean()
```

```python
#Outlairs
fig=sns.scatterplot(df,x="sale_price",y="market_price")
fig.set_title("Outlairs")
```

```python
rating_type=df.groupby(["rating","type"]).size().reset_index(name="count")
```

```python
fig = px.line(rating_type,
            x='rating',
            y='count',
             color='type',
             title='Distribution of  Ratings',
             labels={'rating': 'Rating', 'count': 'Number of Titles', 'type': ' Type'})
fig.show()
```

```python
brand=df.groupby(["brand"]).size().reset_index(name="count")
fig=px.bar(brand.head(20),x="brand",y="count",title="Top 20 Brand",color_discrete_sequence=['orange'])
fig.show()
```

```python
category_counts = df['category'].value_counts().reset_index()
category_counts.columns = ['category', 'count']

fig = px.pie(category_counts,
             values='count',
             names='category',
             title='Distribution of Products by Category')
fig.show()
```
